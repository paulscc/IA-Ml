```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

np.random.seed(42)
tamaño_m2=np.random.randint(50,300,100)
precio=tamaño_m2*2000+np.random.randint(-10000,10000,100)

data ={'tamaño_m2':tamaño_m2, 'precio':precio}
df=pd.DataFrame(data)

X=df[['tamaño_m2']]
y=df[['precio']]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=0)

modelo=LinearRegression()
modelo.fit(X_train,y_train)

predicciones=modelo.predict(X_test)
error=mean_squared_error(y_test, predicciones)
r_error=np.sqrt(error)

print(f"Error cuadratico es:{error}")
print(f"Raiz del error cuadratico es:{r_error}")

#1 Grafico Regresion lineal

# 1. Datos reales y Línea de regresión plt.figure(figsize=(12, 6))
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Datos reales', alpha=0.7)
plt.plot(X, modelo.predict(X), color='red', label='Línea de regresión') 
plt.xlabel('Tamaño en m2')

plt.ylabel('Precio ($)')

plt.title('Relación entre Tamaño y Precio')
plt.legend()
plt.show()
```
