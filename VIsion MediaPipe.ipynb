```python
import cv2
import mediapipe as mp

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
cap = cv2.VideoCapture('Video01.mp4')

contador = 0
arriba = False

def get_y(landmarks, part):
    return landmarks[mp_pose.PoseLandmark[part]].y

def get_styles(up):
    if up:
        return ("SUBEE!!!!", (0, 0, 255),
                mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=3, circle_radius=7),
                mp_drawing.DrawingSpec(color=(0, 0, 255), thickness=5, circle_radius=0))
    else:
        return ("Movimiento correcto", (0, 255, 0),
                mp_drawing.DrawingSpec(color=(0, 0, 255), thickness=3, circle_radius=7),
                mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=5, circle_radius=0))

# Inicia el proceso de pose
with mp_pose.Pose(min_detection_confidence=0.7, min_tracking_confidence=0.7) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        frame = cv2.resize(frame, (800, 800))
        frame = cv2.flip(frame, 1)
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        results = pose.process(rgb)  # Información de landmarks y coordenadas
        
        frame = cv2.cvtColor(rgb, cv2.COLOR_RGB2BGR)  # Convertir frame de vuelta a BGR
        
        if results.pose_landmarks:
            lm=results.pose_landmarks.landmark
            le=get_y(lm,'LEFT_ELBOW')
            lw=get_y(lm,'LEFT_WRIST')
            rw=get_y(lm,'RIGHT_WRIST')
            sh=get_y(lm,'LEFT_SHOULDER')
          
            
            abajo=True

            """
        
            if le< sh and le < sh:  # Ambos codos por encima del hombro (cuando los brazos están extendidos)
                abajo = True  # Significa que los codos están en la parte más alta de la flexión
                status, color, l_style, c_style = get_styles(True)
            elif le > sh and le > sh:  # Ambos codos por debajo del hombro (cuando los codos están doblados)
                if abajo:  # Si antes estábamos arriba, significa que se ha completado una flexión
                    contador += 1  # Contamos una flexión
                    abajo=False  # Ahora cambiamos el estado a "arriba"
                status, color, l_style, c_style = get_styles(False)  # Movimiento hacia abajo
            else:  # Movimiento incompleto, codos no están completamente arriba o abajo
                status, color, l_style, c_style = get_styles(True)  # Movimiento incompleto

            """

            if  sh>0.60:   # Ambos codos por encima del hombro (cuando los brazos están extendidos)
                if abajo:  # Significa que los codos están en la parte más alta de la flexión
                    contador+=0.033
                status, color, l_style, c_style = get_styles(True)
            elif  sh<0.50:  # Ambos codos por debajo del hombro (cuando los codos están doblados)
                if abajo:  # Si antes estábamos arriba, significa que se ha completado una flexión
                     # Contamos una flexión
                    abajo=True  # Ahora cambiamos el estado a "arriba"
                    
                status, color, l_style, c_style = get_styles(False)  # Movimiento hacia abajo
            else:  # Movimiento incompleto, codos no están completamente arriba o abajo
                status, color, l_style, c_style = get_styles(True)  # Movimiento incompleto

          
            
            cv2.putText(frame,f"Flexiones:{round(contador, 0)}\n ",(50,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,0,0),4)
            cv2.putText(frame,status,(25,900),cv2.FONT_HERSHEY_SIMPLEX,2,color,6)
            
           
            mp_drawing.draw_landmarks(
                frame,
                results.pose_landmarks,
                mp_pose.POSE_CONNECTIONS,
                l_style,
                c_style
            )
        
        # Muestra el frame con los landmarks
        cv2.imshow('Landmarks', frame)
        
        # Si se presiona 'q', se detiene el bucle
        if cv2.waitKey(5) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
