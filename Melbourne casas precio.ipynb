```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn. ensemble import RandomForestRegressor

# URL del archivo CSV en la web
url = 'https://raw.githubusercontent.com/itsmecevi/dataset/main/melb_data.csv'

# Leer el CSV directamente desde la URL
df = pd.read_csv(url)
df
# Mostrar las primeras filas del DataFrame
melb_df=df.dropna(axis = 0)
melb_df
y=melb_df['Price']
X=melb_df[['Rooms','Bathroom','Landsize','Lattitude','Longtitude']]

melbourne_model=DecisionTreeRegressor()
melbourne_model.fit(X, y)
```

```python
print("Predicciones 5 casas")
print(X.head())
print("El precio real es")
print(y.head())

print("Las predicciones son")
# Usar el formato adecuado de datos para la predicción

print(melbourne_model.predict(X.head()) ) 
```

```python
predicted_home_prices=melbourne_model.predict(X)
mean_absolute_error(y,predicted_home_prices)
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y)
melbourne_model = DecisionTreeRegressor() 
melbourne_model.fit(train_X, train_y)
val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val) 
    mae
    return(mae)
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(max_leaf_nodes, my_mae))
```

```python

forest_model = RandomForestRegressor() 
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X) 

mean_absolute_error(val_y, melb_preds)
#print (mean_absolute_error(val_y, melb_preds))
```
